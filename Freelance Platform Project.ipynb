```python
def generate_earnings(experience):
    if experience == 'Entry ':
        return np.random.randint(20000, 30000)
    elif experience == 'Intermediate ':
        return np.random.randint(30000, 40000)
    elif experience == 'Expert ':
        return np.random.randint(50000, 60000)
    else:
        return np.nan 
```

```python
df['Earnings'] = df['Experience'].apply(generate_earnings) 
```

```python
df.head()
```

```python
df_pivot = df.pivot_table(values='Earnings', index='Experience', columns='Currency', aggfunc='mean')
fig, ax = plt.subplots()
df_pivot.plot(kind='bar',ax=ax)
ax.set_ylabel('Earnings')
ax.set_xlabel('Experience')
ax.legend(['USD','GBP','EUR']) 
```
